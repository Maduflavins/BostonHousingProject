Attribute Information:

    1. CRIM      per capita crime rate by town
    2. ZN        proportion of residential land zoned for lots over 
                 25,000 sq.ft.
    3. INDUS     proportion of non-retail business acres per town
    4. CHAS      Charles River dummy variable (= 1 if tract bounds 
                 river; 0 otherwise)
    5. NOX       nitric oxides concentration (parts per 10 million)
    6. RM        average number of rooms per dwelling
    7. AGE       proportion of owner-occupied units built prior to 1940
    8. DIS       weighted distances to five Boston employment centres
    9. RAD       index of accessibility to radial highways
    10. TAX      full-value property-tax rate per $10,000
    11. PTRATIO  pupil-teacher ratio by town
    12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks 
                 by town
    13. LSTAT    % lower status of the population
    14. MEDV     Median value of owner-occupied homes in $1000's

# Using RANSAC(RAndom Sapmle Consensus) Algorithm

```python
X = df['RM'].values.reshape(-1,1)
```

```python
y = df['MEDV'].values
```

```python
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor()
```

```python
ransac.fit(X,y)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

```python
line_X = np.arange(3, 10, 1)
line_y_ransac = ransac.predict(line_X.reshape(-1,1))
```

```python
sns.set(style='darkgrid', context='notebook')
plt.figure(figsize=(12,10));
plt.scatter(X[inlier_mask], y[inlier_mask],
           c='blue', marker='o', label='Inliers')
plt.scatter(X[outlier_mask], y[outlier_mask],
           c='brown', marker='s', label='Outliers')
plt.plot(line_X, line_y_ransac, color='red')
plt.xlabel('average number of rooms per dwelling')
plt.ylabel("Median values of owner-occupiedhomes in $1000's")
plt.legend(loc='upper left')
plt.show()
```

```python
ransac.estimator_.coef_
```

```python
ransac.estimator_.intercept_
```

# Performance Evaluation of Regression Model

```python
from sklearn.model_selection import train_test_split
```

```python
X = df['LSTAT'].values.reshape(-1, 1)
```

```python
y = df['MEDV'].values
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_train_pred = lr.predict(X_train)
```

```python
y_test_pred = lr.predict(X_test)
```

# Residual Analysis

```python
plt.figure(figsize=(12,8))
plt.scatter(y_train_pred, y_train_pred-y_train, c='blue', marker='o', label='Training data')
plt.scatter(y_test_pred, y_test_pred - y_test, c='yellow', marker='*', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=10, xmax=50, lw=2, color='k')
plt.xlim([-10, 50])
plt.show()
```

According to the diagram above there are certain datas(outliers) that are not explained by our model

# Mean Squared Errors

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train, y_train_pred)
```

```python
mean_squared_error(y_test, y_test_pred)
```

# Coefficient of Determination

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```
